Durante la etapa previa se identificaron algunos datos faltantes, así como errores en la digitación que dificultan un análisis acertado de los datos, en ese sentido vamos a realizar una serie de operaciones con el fin de preparar el conjunto de datos para su  posterior utilización.

```python
#Limpiando los nombres de los departamentos
MEN_P = MEN_P.replace({"Archipiélago de San Andrés, Providencia y Santa Catalina": "San Andres Islas"})
MEN_P = MEN_P.replace({"Archipiélago de San Andrés. Providencia y Santa Catalina": "San Andres Islas"})
MEN_P = MEN_P.replace({"San Andres": "San Andres Islas"})
MEN_P = MEN_P.replace({"Bogota D.C": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogotá D.C": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogota D.C.": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogotá D.C.": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogota, D.C": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogotá, D.C": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogotá, D.C.": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogotá, D..C": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Bogota": "Bogotá D.C."})
MEN_P = MEN_P.replace({"Norte Santander": "Norte de Santander"})
MEN_P = MEN_P.replace({"Boyaca": "Boyacá"})
MEN_P = MEN_P.replace({"Vaupes": "Vaupés"})
MEN_P = MEN_P.replace({"Bolivar": "Bolívar"})
MEN_P = MEN_P.replace({"Choco": "Chocó"})
MEN_P = MEN_P.replace({"Quindio": "Quindío"})
MEN_P = MEN_P.replace({"Atlantico": "Atlántico"})
MEN_P = MEN_P.replace({"Cesár": "Cesar"})
MEN_P = MEN_P.replace({"Caqueta": "Caquetá"})
MEN_P = MEN_P.replace({"Cordoba": "Córdoba"})
MEN_P = MEN_P.replace({"Guainia": "Guainía"})
MEN_P = MEN_P.replace({"Valle": "Valle del Cauca"})

#Ordenando los datos en orden descendente segun el departamento
MEN_P = MEN_P.sort_values(by=['DEPARTAMENTO'], ascending= False)

#eliminando registros correspondientes al año 2020

MEN_P.drop(MEN_P.loc[MEN_P['AÑO']==2020].index, inplace=True)

#eliminando registros correspondientes al año 2019

MEN_P.drop(MEN_P.loc[MEN_P['AÑO']==2019].index, inplace=True)

#eliminando registros correspondientes al año 2018

MEN_P.drop(MEN_P.loc[MEN_P['AÑO']==2018].index, inplace=True)

#eliminando columna de desercion en transicion que no vamos a usar y posee datos nulos
MEN_P = MEN_P.drop(['DESERCIÓN_TRANSICIÓN'], axis=1)




MEN_P.sample(20)
```

```python
def relacion_indicadores(variable):
  MAT_COR = MEN_P.corr()  
  serieneg = MAT_COR.loc[MAT_COR[variable] < 0, [variable]]*-1  
  seriepos = MAT_COR.loc[MAT_COR[variable] >= 0, [variable]]
  serie = pd.concat([seriepos,serieneg],axis = 0)
  serie = serie.drop(variable)
  serie = serie.sort_values(variable, ascending = False)
  serie2 = serie[variable]
  return serie2

  
```

```python
def relacion_alta_significativa(var_a, var_b, min_mag = 0.5, confianza = 0.95):
  significancia = 1 - confianza
  corr, pvalor = stats.pearsonr(MEN_P[var_a], MEN_P[var_b])  
  if  corr < 0:
    corr = corr * -1
 
  if (pvalor < significancia) & (corr > min_mag):
        es_alta_significativa = True
  else:
        es_alta_significativa = False
  

  return es_alta_significativa
```

**REGRESION LINEAL**

Para esta etapa realizaremos una regresion lineal entre dos indicadores para el departamento de Nariño, con este tipo de estudio se intenta establecer la relacion entre dos variables que crecen o decrecen de manera lineal.

Para nuestro estudio vamos a comparar algunos indicativos reelevantes.

d) Obtenemos los valores correspondientes a la regresion lineal por medio de la funcion lineregress para el departamento que queremos estudiar

e) Generamos la grafica de regresion Lineal


```python
def corr_regr(var_ind, var_dep, departamento):
  ind_df = MEN_P[MEN_P["DEPARTAMENTO"] == departamento]
  crecimiento, valor_inicial, correlación, p_valor, errorStd = stats.linregress(ind_df[var_ind], ind_df[var_dep])
  return crecimiento, valor_inicial, correlación, p_valor
```

```python
var_ind = 'COBERTURA_NETA_PRIMARIA'
var_dep = 'COBERTURA_BRUTA_PRIMARIA'
departamento = 'Nariño'

# NOTA: Su solución debe ser correcta.
m, b, r, p = corr_regr(var_ind, var_dep, departamento) 
mod_lin =  m * MEN_P[var_ind] + b

# Gráfica de dispersión
ax = MEN_P.plot.scatter(var_ind, var_dep, figsize = (6,5)) 
ax.get_figure().set_dpi(105);
ax.plot(MEN_P[var_ind], mod_lin, c='r', label='Regresión lineal / I. Correlacion: '+str(r)); 
ax.legend();
```

Al Observar la regrsion lineal de la cobertura neta en primaria y la cobertura bruta en primaria, para el departamento de nariño, se evidencia la correlacion entre los dos indicadores, creciendo y decreciendo de manera proporcional, en la medida en que la cobertura bruta aumenta, tambien lo hace la cobertura neta primaria, se recomienda al ministerio continuar las politicas y los esfuerzos en aras de aumentar estos indicadores.
